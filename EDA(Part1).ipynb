### How should we replace missing values ???
#### Find relation between independent and dependent feature first

#### Let take all feature if it contain nan put 1 else put 0 then group by 1& 0 and then plot

```python
for feature in features_with_nan:
    data=df.copy()
##    Let take all feature if it contain nan put 1 else put 0 then group by 1& 0 and then plot
    data[feature]=np.where(data[feature].isnull(),1,0)
## Lets calculate the mean sales price where information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature) ### It will show the cloumn name in each graph
    plt.show() ### It will show all graph
```

From the above diagram we conclude that even nan value is missing its sales price(dependent variable) is high.
Eg: For lotfrontage nan value missing(1) have high sales values so we cant eliminate this we have to come up with some
conclusions to replace the valuese through d=feature engineering

Here we will compare all years with salesprice

## EDA Part 2

```python
##We will be using logarithmic transformation(to convet log normal distibution to normal distribution)
```

```python
data=df.copy()
for feature in contineous_Feature:
    if 0 in data[feature].unique():
        pass    ## as log o is infinite
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

## Outliers

```python
for feature in contineous_Feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
